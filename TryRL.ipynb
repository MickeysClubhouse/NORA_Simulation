```python
# OpenAI 
# https://gym.openai.com/docs/

import gym
env = gym.make('CartPole-v0')
env.reset()
for _ in range(1000):
    #env.render() # it seems jupyter notebook cannot render this, but pycharm will work
    env.step(env.action_space.sample()) # take a random action
env.close()
```
